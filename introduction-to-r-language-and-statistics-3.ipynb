```r
library(wordcloud)
```

```r
library(tm)
```

```r
dirPath = "/kaggle/input/simple-dataset-text"

corpus <- Corpus(DirSource(dirPath))
```

```r
inspect(corpus)
```

## Data Cleaning

```r
corpus <- tm_map(corpus, content_transformer(tolower))
```

```r
corpus <- tm_map(corpus, removeNumbers)
```

```r
corpus <- tm_map(corpus, removeWords, stopwords("english"))
```

```r
corpus <- tm_map(corpus, removeWords, c("floccinaucinihilipification", "squirrelled"))
```

```r
corpus <- tm_map(corpus, removePunctuation)
```

```r
corpus <- tm_map(corpus, stripWhitespace)
```

```r
corpus
```

```r
# Create a Term Document Matrix
dtm <- TermDocumentMatrix(corpus)
```

```r
dtm
```

```r
# Matrix transformation
m <- as.matrix(dtm)
m
```

```r
# Sort it to show the most frequent words
v <- sort(rowSums(m), decreasing=TRUE)
v
```

```r
# Transform to a data frame
d <- data.frame(word=names(v), freq=v)
head(d, 10)
```

```r
wordcloud(words = d$word, freq=d$freq)
```

```r
wordcloud(words = d$word, freq=d$freq, min.freq=1)
```

```r
wordcloud(words = d$word, freq=d$freq, min.freq=1, max.words=250, colors=brewer.pal(8, "Dark2"))
```

```r
wordcloud(words = d$word, freq=d$freq, min.freq=1, max.words=250, colors=brewer.pal(8, "Dark2"), random.order=FALSE)
```
